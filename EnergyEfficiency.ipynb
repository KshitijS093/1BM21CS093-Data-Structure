```python
import pandas as pd
import numpy as np
df = pd.read_csv("/content/IOT-temp.csv")
```

```python
df.head()
```

```python
df=df.filter(items=['noted_date','temp','out/in'])
```

```python
df=df.drop(df[df['out/in']=='In'].index)
#removes the indoor temperature values as they are less in number
```

```python
df['noted_date'] = pd.to_datetime(df['noted_date'])
df['Day'] = df['noted_date'].dt.day
df['Month']=df['noted_date'].dt.month
df['Hour'] = df['noted_date'].dt.hour
df['Minute']=df['noted_date'].dt.minute
#here we converted it to date time format and then extracted date and time from the respective column
```

```python
df=df.filter(['Day','Month','Hour','Minute','temp'])
```

```python
df.isnull().sum()
#hence no null values are present in the dataset that contains values of indoor temperature
```

```python
df=df.drop_duplicates()
```
